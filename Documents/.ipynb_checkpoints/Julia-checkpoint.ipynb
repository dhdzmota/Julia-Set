```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constant complex number
c=complex(-0.162, 1.04)

#Threshold value of the magnitude.
z_th=1000

# Maximum number of iterations
max_iteration=50

# Plane delimitations:

x_min = -1.5
x_max =  1.5

y_min = -2
y_max =  2

# Width and height of the picture (matrix that represents the
# complex plane) 
w = 1000
h = 1000

# Lenght of each side of the plane
xw = x_max - x_min
yw = y_max - y_min

# Creation of the complex plane
complex_plane = np.zeros((h, w))

# Filling of the values
for i in range(w):
    for j in range(h):
        iteracion = 0
        z = complex(i/w*xw + x_min, j/h*yw + y_min)

        #Iterations
        while abs(z) <= z_th and iteracion < max_iteration:
            z = z**2 + c
            iteracion += 1
        ratio = iteracion / max_iteration
        complex_plane[j,i] = ratio
```

```python
fig = plt.subplots(figsize=(20.1,10.))
plt.imshow(complex_plane, interpolation='quadric', cmap='gist_yarg')
if np.imag(c) < 0:
    plt.title(f'Julia set: c = {np.real(c)}-{abs(np.imag(c))}i, f(z)=')
elif np.imag(c)>0:    
    plt.title(f'Julia set: c = {np.real(c)}+{np.imag(c)}i, f(z)=')
else:
    plt.title(f'Julia set: c = {np.real(c)}, f(z)=')

plt.axis('off')
plt.savefig('c*cos(z)*sin(z2)+c.png')
plt.show()
```
